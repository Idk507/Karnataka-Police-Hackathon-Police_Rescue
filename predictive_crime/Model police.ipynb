```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataset is stored in a DataFrame called 'data'
# and it contains columns like 'Year', 'Month', and 'CrimeGroup_Name'

# Replace 'data' with the actual name of your DataFrame

# Get unique years present in the dataset
years = [2016,2017,2018,2019,2020,2021,2022,2023,2024]

# Loop through each year and plot top 5 crime categories occurrences by month
for year in years:
    # Filter data for the current year
    crime_data_year = data[data['Year'] == year]
    try:
        # Group data by month and count occurrences of each crime category
        monthly_crime_counts = crime_data_year.groupby(['Month', 'CrimeGroup_Name']).size().unstack(fill_value=0)
        
        # Get the total occurrences of each crime category across all months
        total_crime_counts = monthly_crime_counts.sum()
        
        # Get the top five crime categories
        top_5_crime_categories = total_crime_counts.nlargest(5).index
        
        # Filter the monthly crime counts to include only the top five crime categories if they exist
        top_5_monthly_crime_counts = monthly_crime_counts[top_5_crime_categories] if len(top_5_crime_categories) > 0 else pd.DataFrame()
        
        # Reindex the DataFrame with all months
        all_months = range(1, 13)
        top_5_monthly_crime_counts = top_5_monthly_crime_counts.reindex(all_months, fill_value=0)
        
        # Plotting if there are any top 5 crime categories
        if not top_5_monthly_crime_counts.empty:
            top_5_monthly_crime_counts.plot(kind='bar', figsize=(12, 8), width=0.8)
            plt.title(f'Top 5 Crime Categories Occurrences by Month - {year}')
            plt.xlabel('Month')
            plt.ylabel('Number of Occurrences')
            plt.xticks(range(len(all_months)), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
            plt.legend(title='Crime Category')
            plt.tight_layout()
            plt.show()
        else:
            print(f"No top 5 crime categories found for the year {year}.")
    except KeyError:
        print(f"No data available for the year {year}.")
```

```python
# Adding cluster labels to the DataFrame
crime_data_cluster_encoded['Cluster'] = clusters
```
